```python
from pycaret.datasets import get_data
dataset = get_data('data_train')
```

```python
train = dataset.sample(frac=0.90, random_state=786) # 데이터 분리
test = dataset.drop(train.index)
train.reset_index(inplace=True, drop=True)
test.reset_index(inplace=True, drop=True)
```

```python
from pycaret.classification import *
```

```python
exp_clf = setup(data = train, target = 'weather label', session_id=123, use_gpu = True)
```

```python
best_model1 = compare_models()
```

```python
rf = create_model('rf')
final_model = finalize_model(rf)
prediction = predict_model(rf, data = test)
prediction = predict_model(final_model, data = test)
```

```python
tuned_rf = tune_model(rf)
final_model = finalize_model(tuned_rf)
prediction = predict_model(tuned_rf, data = test)
prediction = predict_model(final_model, data = test)
```

```python
catboost = create_model('catboost')
final_model = finalize_model(catboost)
prediction = predict_model(catboost, data = test)
prediction = predict_model(final_model, data = test)
tuned_catboost = tune_model(catboost)
final_model = finalize_model(tuned_catboost)
prediction = predict_model(tuned_catboost, data = test)
prediction = predict_model(final_model, data = test)
```
